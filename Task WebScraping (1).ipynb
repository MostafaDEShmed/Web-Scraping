```python
import csv
import requests
from bs4 import BeautifulSoup
import json
```

```python
url = 'https://www.baraasallout.com/test.html'
response = requests.get(url)
print(response.text)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
headers_1 = soup.find_all('h1')
headers_1
```

```python
all_p_li = soup.find_all(['p', 'li'])
all_p_li
```

```python
with open('Extract_Text_Data.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Tag", "Content"])
    writer.writerows(headers_1 + all_p_li)
```

```python
with open('products.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Product Name", "Price", "Stock Status"])  

    for row in soup.find_all('tr'):
        name = row.find('td', class_='product-name')  
        price = row.find('td', class_='product-price')  
        stock = row.find('td', class_='stock-status')  

        if name and price and stock:
            writer.writerow([name.get_text(strip=True), price.get_text(strip=True), stock.get_text(strip=True)])
```

```python
books = []
for card in soup.find_all('div', class_='book-card'):  
    title = card.find('h3').text.strip()
    price = card.find('span', class_='price').text.strip()
    stock = card.find('span', class_='stock').text.strip()
    button = card.find('button').text.strip()

    books.append({
        "Book Title": title,
        "Price": price,
        "Stock Availability": stock,
        "Button Text": button
    })


```

```python
with open('books.json', 'w', encoding='utf-8') as file:
    json.dump(books, file, ensure_ascii=False, indent=4)
```

```python
with open('books.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=["Book Title", "Price", "Stock Availability", "Button Text"])
    writer.writeheader()
    writer.writerows(books)
```

```python
inputs = []

for input_tag in soup.find_all('input'):
    inputs.append({
        "Field Name": input_tag.get('name'),
        "Input Type": input_tag.get('type'),
        "Default Value": input_tag.get('value', '')
    })

with open('input_fields.json', 'w', encoding='utf-8') as file:
    json.dump(inputs, file, indent=4)
```

```python

data = []

for a_tag in soup.find_all('a', href=True):
    data.append({
        "Link Text": a_tag.get_text(strip=True),
        "URL": a_tag['href']
    })

for iframe_tag in soup.find_all('iframe', src=True):
    data.append({
        "Video URL": iframe_tag['src']
    })

with open('links_and_videos.json', 'w', encoding='utf-8') as file:
    json.dump(data, file, indent=4)
```
